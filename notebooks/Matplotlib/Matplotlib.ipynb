```python
from scipy import pi
import numpy as np
x = np.linspace(0,4*pi,100)
y = np.cos(x)
plt.plot(x,y,"co:")
```

```python
plt.plot(x,y, x,y**2, x,y*4)
plt.xlabel("aaa")
plt.ylabel("aaa")
plt.title("aaa")
plt.savefig("grafico1.png", transperant = True)
```

```python
plt.plot(x,y, x,y**2, x,y*4)
plt.xlabel("aaa")
plt.ylabel("aaa")
plt.title("aaa")
plt.tight_layout()
plt.savefig("grafico2.png", transperant = True)
```

```python
fig, ax = plt.subplots()
ax.plot([0,3,5,7,8,9], [0, 5, 8, 9, 6, 7], "o--r", linewidth = 3, label = "linha vermelha")
ax.plot([1,2,3,4,5,6], [5,7,9,3,8,5], color = "#444444", marker = "o", linestyle = ":", linewidth=1, label = "linha cinza")
ax.set_xlabel("eixo X")
plt.legend()
```

```python
fig, ax = plt.subplots(1,2, sharey = True, sharex = True, figsize = (5,4))

ax[0].plot([0,3,5,7,8,9], [0, 5, 8, 9, 6, 7], "o--r", linewidth = 3, label = "linha vermelha")
ax[0].plot([1,2,3,4,5,6], [5,7,9,3,8,5], color = "#444444", marker = "o", linestyle = ":", linewidth=1, label = "linha cinza")
ax[0].set_xlabel("eixo x")
ax[0].legend()
ax[1].plot([0,3,5,7,8,9], [0, 5, 8, 20, 6, 7], "o--r", linewidth = 3, label = "linha vermelha")
ax[1].plot([1,2,3,4,5,6], [5,7,9,3,8,5], color = "#444444", marker = "o", linestyle = ":", linewidth=1, label = "linha cinza")
ax[1].set_xlabel("EIXO X")
ax[1].legend()
plt.tight_layout()
```

```python
import pandas as pd
iris = pd.read_csv("/home/elias/estudandopython/datasets/iris.csv")
iris
```

```python
color = {"setosa":"orange","versicolor":"blue","virginica":"black"}

iris["color"] = iris["species"].map(color)

fig, ax = plt.subplots()

ax.scatter(iris["sepal_length"], iris["sepal_width"], c = iris["color"])

#ax.scatter(iris.iloc[0:50,0], iris.iloc[0:50,1])
#ax.scatter(iris.iloc[50:100,0], iris.iloc[50:100,1])
#ax.scatter(iris.iloc[100:150,0], iris.iloc[100:150,1])

ax.set_xlabel("Comprimento das Sépalas")
ax.set_ylabel("Largura das Sépalas")
ax.set_title("Comprimento X Largura das Sépalas")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, sharey = True, sharex = True, figsize = (8,5))
fig.suptitle("Comprimento X Largura")
ax[0].scatter(iris.iloc[0:50,0], iris.iloc[0:50,1], label = "Setosa")
ax[0].scatter(iris.iloc[50:100,0], iris.iloc[50:100,1], label = "Versicolor")
ax[0].scatter(iris.iloc[100:150,0], iris.iloc[100:150,1], label = "Virginica")
ax[0].set_xlabel("Comprimento das Sépalas")
ax[0].set_ylabel("Largura das Sépalas")
ax[1].scatter(iris.iloc[0:50,2], iris.iloc[0:50,3])
ax[1].scatter(iris.iloc[50:100,2], iris.iloc[50:100,3])
ax[1].scatter(iris.iloc[100:150,2], iris.iloc[100:150,3])
ax[1].set_xlabel("Comprimento das Pétalas")
ax[1].set_ylabel("Largura das Pétalas")
ax[0].legend()
plt.tight_layout()
```
